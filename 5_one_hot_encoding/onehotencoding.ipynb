```python
import pandas as pd 
import numpy as np
import sklearn
```

```python
df=pd.read_csv("carprices.csv")
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.xlabel('milage')
plt.ylabel('age')
plt.scatter(df['Mileage'],df['Age(yrs)'],color='red',marker='o')
```

```python
dummies=pd.get_dummies(df['Car Model'])
dummies
```

```python
mer=pd.concat([df,dummies],axis='columns')
mer
```

```python
x=mer.drop(['Car Model','Sell Price($)','Mercedez Benz C class'],axis='columns')
x
```

```python
y=mer['Sell Price($)']
y
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
```

```python
 model.fit(x,y)
    
```

```python
model.score(x,y)
```

```python
model.predict([[45000,4,0,0]])
```
