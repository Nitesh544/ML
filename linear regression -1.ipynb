```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import linear_model 
```

```python
df=pd.read_csv('cpi.csv')
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('cpi')
plt.scatter(df.year,df.percapitaincome,color='blue',marker='+')
```

```python
ydf=df.drop('percapitaincome',axis='columns')
ydf
```

```python
cpi=df.percapitaincome
cpi
```

```python
lg=linear_model.LinearRegression()
lg.fit(ydf,cpi)
```

```python
lg.predict([[50]])
```

```python
lg.coef_
```
